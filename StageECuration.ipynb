```python
#importing necessary libraries
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading partly edited datasets from excel worksheets
d2009 = pd.read_excel("2010.xlsx", sheet_name='d2009', index_col='Features')
d2010 = pd.read_excel("2010.xlsx", sheet_name='d2010', index_col='Features')
d2011 = pd.read_excel('2011Q1.xlsx', sheet_name='Q1', index_col='Features')
d2012 = pd.read_excel('2012Q.xlsx', sheet_name='Q2012', index_col='Features')
d2013 = pd.read_excel('2013Q.xlsx', sheet_name='F2013', index_col='Features')
d2014 = pd.read_excel('2014Q1.xlsx', sheet_name='F2014', index_col='Features')
d2015 = pd.read_excel('2015Q1.xlsx', sheet_name='F2015', index_col='Features')
d2016 = pd.read_excel('2016Q1.xlsx', sheet_name='F2016', index_col='Features')
d2017 = pd.read_excel('2017Q1.xlsx', sheet_name='F2017', index_col='Features')
d2018 = pd.read_excel('2018Q1.xlsx', sheet_name='F2018', index_col='Features')
d2019 = pd.read_excel('2019Q1.xlsx', sheet_name='F2019', index_col='Features')
```

```python
#removing index name
d2009.index.name = None
d2010.index.name = None
d2011.index.name = None
d2012.index.name = None
d2013.index.name = None
d2014.index.name = None
d2015.index.name = None
d2016.index.name = None
d2017.index.name = None
d2018.index.name = None
d2019.index.name = None
```

```python
#transposing datasets for easy merging
d2009 = d2009.T
d2010 = d2010.T
d2011 = d2011.T
d2012 = d2012.T
d2013 = d2013.T
d2014 = d2014.T
d2015 = d2015.T
d2016 = d2016.T
d2017 = d2017.T
d2018 = d2018.T
d2019 = d2019.T
```

```python
#converting 2009 dataset to float type
d2009 = d2009.astype(float)
d2009.info()
```

```python
#Concatenation
pd.concat([d2009, d2010, d2011, d2012, d2013, d2014, d2015, d2016, d2017, d2018, d2019], 
          keys = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019], 
         names=['Year', 'Quarter'])
```

```python
#assigning combines dataset to a new df
finance = pd.concat([d2009, d2010, d2011, d2012, d2013, d2014, d2015, d2016, d2017, d2018, d2019], 
          keys = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019], 
         names=['Year', 'Quarter'])
finance.info()
```

```python
#checking the percentage null value
finance.isnull().mean() * 100
```

```python
finance.columns
```

```python
#dropping columns above 50% missing values
finance = finance.drop(columns=['Discontinued Operation, Tax Effect of Discontinued Operation',
       'Income (Loss) from Continuing Operations, Net of Tax, Including Portion Attributable to Noncontrolling Interest',
       'Income (Loss) from Discontinued Operations, Net of Tax, Including Portion Attributable to Noncontrolling Interest',
       'Income (Loss) from Continuing Operations, Per Basic Share',
       'Discontinued Operation, Income (Loss) from Discontinued Operation, Net of Tax, Per Basic Share',
       'Income (Loss) from Continuing Operations, Per Diluted Share',
       'Discontinued Operation, Income (Loss) from Discontinued Operation, Net of Tax, Per Diluted Share'])
```

## Visualizations

```python
#trends per year
plt.figure(figsize=(20,10))
finance.loc[(2009)].drop('Year ending')['NET OPERATING REVENUES'].plot(kind='bar')
plt.title('Growth trends of 2009 revenues BY Quarter')
plt.show()
```

Second quarter appears to have the highest revenue in 2009

```python
plt.figure(figsize=(20,10))
finance['NET OPERATING REVENUES'].plot(kind='bar')
plt.title('Net revenue trends by quarter by year')
plt.show()
```

Each year seems to follow the same trend of high net revenues in the second quarter. The year 2021 has the highest net revenue over 11 years of financial record.

```python
plt.figure(figsize=(20,10))
finance.loc[(2009)].drop('Year ending')['GROSS PROFIT'].plot(kind='bar')
plt.title('Growth trends of 2009 Profits BY Quarter')
plt.show()
```

```python
plt.figure(figsize=(20,10))
finance['GROSS PROFIT'].plot(kind='bar')
plt.title('Gross profit trends by quarter by year')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Year', y='NET INCOME ATTRIBUTABLE TO SHAREOWNERS OF THE COCA-COLA COMPANY', hue='Quarter', data=finance, legend = 'full')
plt.show()
```

In the history of Coca cola, 2010 recorded the highest net income attributable to shareholders

```python
plt.figure(figsize=(20,10))
finance['Selling, general and administrative expenses'].plot(kind='bar')
plt.title('Expenses trends by quarter by year')
plt.show()
```
